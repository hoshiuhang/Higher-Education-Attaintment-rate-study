## import enrolment ratio file

```python
enrolment_ratio = pd.read_csv('SDG_4.3.2_Asia.csv')
```

```python
enrolment_ratio = enrolment_ratio[['Indicator','Country','Time','Value']]
enrolment_ratio.head()
```

```python
enrolment_ratio.Indicator.unique()
```

## *** enrolment_ratio

```python
enrolment_ratio = enrolment_ratio.loc[enrolment_ratio['Indicator']=='Gross enrolment ratio for tertiary education, both sexes (%)']
enrolment_ratio
```

## Time series

### create a time range

```python
time = []
time_value = 1970
time_max = 2020
while time_value <= time_max:
    time.append(time_value)
    time_value += 1
```

### convert it to a DataFrame

```python
time
time_range = DataFrame(time,columns=['year'])
```

```python
time_range
```

## Right join enrolment ratio and time range 

```python
df = pd.merge(enrolment_ratio,time_range,left_on='Time',right_on='year',how='right')
```

```python
df.sort_values(by=['Country'])
```
