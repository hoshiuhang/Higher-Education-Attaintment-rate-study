## *** enrolment_ratio

```python
enrolment_ratio = enrolment_ratio.loc[enrolment_ratio['Indicator']=='Gross enrolment ratio for tertiary education, both sexes (%)']
enrolment_ratio
```

## check the country on both list

```python
edu_exp_percent_country = set(edu_exp_percent.Country.unique()) #countries in education expend
edu_exp_percent_country
```

```python
enrolment_ratio_country = set(enrolment_ratio.Country.unique()) #countries in enrolment rate
enrolment_ratio_country
```

```python
difference = enrolment_ratio_country.difference(edu_exp_percent_country)
difference
```

### check the defference between 2 datasets

```python
# prints the missing and additional elements in enrolment_ratio_country 
print("Missing values in enrolment_ratio_country list:", (set(edu_exp_percent_country).difference(enrolment_ratio_country))) 
print("Additional values in enrolment_ratio_country list:", (set(enrolment_ratio_country).difference(edu_exp_percent_country))) 
  
# prints the missing and additional elements in list1 
print("Missing values in edu_exp_percent_country list:", (set(enrolment_ratio_country ).difference(edu_exp_percent_country))) 
print("Additional values in edu_exp_percent_country list:", (set(edu_exp_percent_country).difference(enrolment_ratio_country ))) 
```

## Inner join two list together

```python
merged_df = enrolment_ratio.merge(edu_exp_percent, how='inner', left_on=["Country", "Time"], right_on=["Country","Time"])
merged_df
```

```python
type(edu_exp_percent)
```

```python
merged_df.rename(columns={'Value_x':'enrolment_rate','Value_y':'edu_exp_%'}, inplace=True)
merged_df.head()
```

## calculate  the corelation

```python
cor_enrol_edu = merged_df.groupby('Country')[['enrolment_rate','edu_exp_%']].corr().unstack().iloc[:,1]
cor_enrol_edu
```

```python
df = cor_enrol_edu.to_frame()
df
```

```python
type(df)
```

```python
df.columns = ['correlation']
edu_correlation = df.sort_values(by=['correlation'], ascending=False)
edu_correlation
```

## count number of record by country

```python
no_of_record = merged_df.groupby('Country')[['Country']].count()
#no_of_record
#type(no_of_record)

no_of_record.columns = ['record']
no_of_record
```

```python
#merged_df = enrolment_ratio.merge(edu_exp_percent, how='inner', left_on=["Country", "Time"], right_on=["Country","Time"])
cor_exp = edu_correlation.merge(no_of_record, how='inner', left_on=['Country'], right_on=['Country'])
cor_exp
```
